```python
import pandas as pd
```

```python
dataset = pd.read_csv("Banking Mobile App Usage Survey - English.csv")
```

```python
dataset.head()
```

```python
import seaborn as sns
```

```python
sns.catplot(x="What is your gender?", kind="count", palette="Blues", data=dataset)
```

```python
g =sns.catplot(x="Which age group do you belong to?", kind="count", palette="Blues",
            order=["Under 16 years", "16 - 25","25 - 45","45 - 60","60 or above"],data=dataset)
g.set_xticklabels(rotation=30)
```

```python
g =sns.catplot(x="What district do you currently live in?", kind="count", palette="Blues",
            data=dataset, order = dataset['What district do you currently live in?'].value_counts().index)
g.set_xticklabels(rotation=90)
```

```python
g =sns.catplot(x="Marital status ?", kind="count", palette="Blues",
            data=dataset, order = dataset['Marital status ?'].value_counts().index)
g.set_xticklabels(rotation=30)
```
